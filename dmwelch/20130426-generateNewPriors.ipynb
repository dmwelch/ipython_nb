```python
files = ['summed_1457/summed_1457_POSTERIOR_VB',
         'summed_1457/summed_1457_POSTERIOR_NOTVB',
         'globus', 
         'thalamus', 
         'hippocampus', 
         'basal_ganglia', 
         'out_gm', 
         'soft_POSTERIOR_SURFGM',
         'summed_1457/summed_1457_POSTERIOR_NOTGM',
         'out_wm', 
         'soft_POSTERIOR_WM', 
         'summed_1457/summed_1457_POSTERIOR_NOTWM',
         'summed_1457/summed_1457_POSTERIOR_CSF',
         'summed_1457/summed_1457_POSTERIOR_NOTCSF']
imgs = []
c = 0
for filename in files:
    full = os.path.join(experimentDir, filename + ".nii.gz")
    assert os.path.exists(full), full
    imgs.append(sitk.ReadImage(full)) # rounded)
```

```python
def clampImage(image, lowerThreshold=0.001, lowerValue=0.0, upperThreshold=0.999, upperValue=1.0):
    stats = sitk.Statistics(image)
    image_max = stats["Maximum"]
    lowered = sitk.Threshold(image, lower=lowerThreshold, upper=image_max, outsideValue=lowerValue)
    l_stats = sitk.Statistics((lowered >= lowerValue) - (lowered >= lowerThreshold))
    try:
        assert l_stats["Sum"] == 0, "There are values between %g and %g!" % (lowerValue, lowerThreshold)
    except AssertionError, e:
        sitk.WriteImage(lowered, os.path.join(experimentDir, "clampImage_lowered.nii.gz"))
        sitk.WriteImage((lowered > lowerValue) - (lowered >= lowerThreshold), os.path.join(experimentDir, "clampImage_lowerMask.nii.gz"))
        raise e
    clamped = sitk.Threshold(lowered, lower=lowerValue, upper=upperThreshold, outsideValue=upperValue) 
    c_stats = sitk.Statistics((clamped < upperValue) - (clamped <= upperThreshold))
    assert c_stats["Sum"] == 0, "There are values between %g and %g!" % (upperValue, upperThreshold)
    return clamped
        
def truncate(image, ndigits=0):
    """ Truncate to the given number of digits """
    from decimal import Decimal
    import numpy as np
    narray = sitk.GetArrayFromImage(image)
    rounded = np.around(narray, decimals=ndigits)
    truncated = sitk.GetImageFromArray(rounded)
    truncated.CopyInformation(image)
    return truncated
```
