```python
def filesplit(filename):
    """ 
    Split filename by the first '.' in order to append to filename (not extension)
    """
    flist = filename.split('.')
    if len(flist) == 2:
        return flist[0], '.' + flist[1]
    else:
        return flist[0], '.' + '.'.join(flist[1:])
```

```python
def resampleImage(infile, outfile, resolution=(1.0, 1.0, 1.0), interpolator=sitk.sitkLinear):
    image = sitk.ReadImage(infile)
    outImage = image
    old_size = list(image.GetSize())
    old_res = list(image.GetSpacing())
    new_res = list(resolution)
    new_size = [ int((old_size[i] * old_res[i]) / new_res[i]) for i in range(len(new_res))]
    outImage = sitk.Resample(outImage,
                             transform=sitk.Transform(),
                             outputSpacing=resolution,
                             interpolator=interpolator,
                             size=tuple(new_size),
                             outputOrigin=image.GetOrigin(),
                             outputDirection=image.GetDirection())
    felements = outfile.split('.')
    if len(felements) == 1:
        outfile += '.nii.gz'
    outfile = os.path.join(os.getcwd(), outfile)
    sitk.WriteImage(outImage, outfile)
    return outfile
```

## Create Brain mask ##

We want to do several steps:

1. Threshold the probability images at 0.1
1. Mask the TISSUE files with a ICV or Brain label 
1. Dilate the mask by a width of 3 mm  
1. Blur the mask by a Gaussian with a kernel twice the width of the dilation.  
1. The TISSUE image will be multiplied by the blurred mask, and 
1. the NOT_TISSUE image will be multiplied by (1 - blurred_mask).  

This should allow us to iterate over all voxels of overlap to verify that P(tissue) + P(not) = 1.0

### Method 1 ###
This method was reasonable, but the NAC atlas has an artifact at the right posterior occipital hemisphere notch - the extent of the brain mask is hampered by the VB prefering the right side.

## A better brain mask ##

1.  Threshold all tissue probabilities by 0.01 (1%)
2.  Sum all the tissue probabilities 
3.  Create a binary mask at 0.5 (50%)?

```python
def SoftMask(image, FWTM, directions=[0,1,2]):
    """ 
    Returns the Gaussian blurred (soft) mask of a binary (hard) mask with the full-width-tenth-max (FWTM) width 
    """
    import math
    factor = 2 * math.sqrt(2 * math.log(10))
    sigma = FWTM / factor
    soft = image
    for direction in directions:
        soft = sitk.RecursiveGaussian(soft, sigma=sigma, direction=direction)
    return soft
```

```python
def clampImage(image, lowerValue=0.0, upperValue=1.0):
    array = sitk.GetArrayFromImage(image)
    clamped_array = numpy.where(array > upperValue, numpy.ones(array.shape) * upperValue, array)
    clamped_array = numpy.where(clamped_array < lowerValue, numpy.ones(array.shape) * lowerValue, clamped_array)
    clamped_image = sitk.GetImageFromArray(clamped_array)
    clamped_image.CopyInformation(image)
    return clamped_image
```
