# Subject-specific prior corrections #

We want to run our algorithms on each subject seperately, instead of as a group.  Use the files in /paulsen/Experiments/20130202_PredictHD_Results/SubjectToAtlasWarped/*/

1. Fix the air
  1. Due to a registration error, the default value for the outside value was set to zero (this is right for ALL the other images!), but for AIR the image background is the area of interest in all the other images
$$
AIR = AIR_0 + (1 - \sum(POSTERIORS))
$$
1. Create the basal ganglia
  1. Sum accumben, caudate, and putamen masks
  1. Create a binary mask for sum > 0 and save as "cleaned_basal_ganglia.nii.gz"
1. Correct the gray matter
  1. <strike>For each mask pair (right, left): </strike>
    1.  <strike>Sum the mask pair</strike>
    1.  <strike>Mask the corresponding POSTERIOR_* file with the inverse binary mask of the sum mask</strike>
    1.  <strike>Add the resulting image to POSTERIOR_SURFGM.nii.gz</strike>
  1. Add all POSTERIOR images for SURFGM
  1. Mask the summed SURFGM with the sum of all masks
1. Remove mask residuals
  1. Create an inverse binary mask from mask images ("cleaned_*")
  1. Finally, mask all the POSTERIOR_* files for normalization, _(even SURFGM!)_ by the inverse mask (e.g. don't use POSTERIOR_PUTAMEN in normalization)
1. Normalize and sum
  1. Normalize the images s.t. all voxels in the sum equal 1


## Correct the air ##

```python
def sumPosteriors(inputDict):
    airKeys = [key for key in inputDict.keys() if key[1] != "_"]  # Get all the keys that **aren't** a segmentation
    posteriorSum = sumOverKeys(inputDict, airKeys)
    return posteriorSum

def correctAir(air, summed, show=False):
    """ Given the air image and the summed image of all posteriors, correct the air values lying on the periphery and return the image """
    sum_mask = sitk.BinaryThreshold(summed, lowerThreshold=1.0, insideValue=0, outsideValue=1)
    if show: myshow(sum_mask, "Mask of summed @ 1.0")
    relabel_mask = sitk.RelabelComponent(sitk.ConnectedComponent(sum_mask))
    if show: myshow(sitk.LabelToRGB(relabel_mask), "Relabeled mask")
    clean_mask = relabel_mask==1
    if show: 
        myshow(clean_mask, "Cleaned mask")
        myshow(air, "Original air")
    fixed_air = air + sitk.Cast(clean_mask, air.GetPixelIDValue())
    open_mask = sitk.BinaryMorphologicalOpening(fixed_air>0.01, 3)
    fixed_air = sitk.Mask(fixed_air, open_mask)
    if show: myshow(fixed_air, "Fixed air")
    return fixed_air

airsum = sumPosteriors(inputDict)
# myshow(inputDict['air'][:,182,:], "original")
out1 = {}
out1['air'] = correctAir(inputDict['air'], airsum, False)
# myshow(out1['air'][:,182,:], "corrected")
if debug:
    checkStats(out1, checks=["min"])

for key in inputDict.keys():
    if key in out1.keys():
        continue
    else:
        out1[key] = inputDict[key]
inputDict = None
```

```python
def createResidualImage(d, ikeys, mkeys):
    """
    For a list of images and masks, sum both groups (summed image and summed mask) and return the summed image 
    masked with the NOT summed mask
    """
    invert_mask = sumOverKeys(d, mkeys) < 1
    res_sum = sumOverKeys(d, ikeys)
    # myshow(sitk.LabelOverlay(res_sum, invert_mask), "Masked SurfGM")
    residual = sitk.Mask(res_sum, invert_mask)
    # myshow(residual, "SurfGM w/ Residuals")
    # assert sitk.Statistics(residual)["Maximum"] > 0.0, "residual is empty for %s" % orig
    return residual

gmImageKeys = ['surfgm', 'crblgm', 'notgm',   # posterior gm
             'accumben', 'caudate', 'globus', 'hippo', 'putamen', 'thalamus']  # posterior

maskKeys = ['mask_basal',
            'mask_globus',
            'mask_hippo', 
            'mask_thalamus']
out2['surfgm'] = createResidualImage(out2, gmImageKeys, maskKeys)
if debug:
    checkStats(out2, checks=['min'])
```

# BRAINSTEM #

The WM probabilities in the brainstem are in the cerebellum map, NOT the cerebral map.

1. Create a brainstem mask
2. Dilate by 1mm
3. Sum (mask * cerebellum) + cerebral map = new cerebral map
4. Sum (1 - mask) * cerebellum map = new cerebellum map

NOTE: This is done BEFORE cerebellar masking of the cerebellar WM!

Width of Gaussian:

To get the width of the blurred edge at $1/10$ of maximum (FWTM): 

$$FWTM = 2\sigma\sqrt{2\ln(10))}$$
$$\sigma = {{FWTM}\over{(2\sqrt{2\ln(10)}}}$$

```python
def SoftMask(image, x, method='FWTM'):
    """ 
    Returns the Gaussian blurred (soft) mask of a binary (hard) mask with either:
        * full-width-approx-zero (FWAZ) width 
        * full-width-tenth-max (FWTM) width (default)
        * full-width-half-max (FWHM) width

    Note: FWTM has been found to be the most intuitive of the three, giving a 'noticible' blur for 1 mm
    """
    import math
    if method.lower() == "fwtm":
        # print "FWTM"
        z = 10.0
    elif method.lower() == "fwaz":
        # print "FWAZ"
        z = 100.0
    elif method.lower() == "fwhm":
        # print "FWHM"
        z = 2.0
    factor = 2.0 * math.sqrt(2.0 * math.log(z))
    sigma = float(x) / factor
    # print sigma
    variance = sigma ** 2
    if image.GetPixelIDValue() != sitk.sitkUInt8:
        raise IOError
    soft = sitk.DiscreteGaussian(sitk.Cast(image, sitk.sitkFloat32), variance=float(variance))
    return soft

maskDict = createImageMap(brainstem=osjoin(experimentDir, 'newAtlas', 'brainstem.nii.gz'),
                          cerebrum=osjoin(experimentDir, 'newAtlas', 'cerebrum.nii.gz'),
                          cerebellum=osjoin(experimentDir, 'newAtlas', 'cerebellum.nii.gz'))
# resample masks to prior space
mask1 = {}
for key, value in maskDict.items():
    mask1[key] = sitk.Resample(maskDict[key],
                               interpolator=sitk.sitkNearestNeighbor,
                               size=maskDict[key].GetSize(),
                               outputOrigin=out3['air'].GetOrigin(),
                               outputDirection=out3['air'].GetDirection())
assert mask1.keys() == maskDict.keys()

# Create hard and soft masks
# combine the cerebrum and brainstem mask
mask1['cerebrum'] = (mask1['cerebrum'] + mask1['brainstem']) > 0
mask1.pop('brainstem')
# myshow(mask1['cerebrum'], title="Original")
# myshow(mask1['cerebellum'][:,:,96], title="Original")
if debug:
    checkStats(mask1)
maskDict = None

mask2 = {}
for key in mask1.keys():
    # Create hard mask
    mask2[key] = sitk.BinaryMorphologicalClosing(mask1[key], 3) # close hard masks
    #if key == 'cerebellum':
    #    myshow((mask2[key] - mask1[key])[:,:,96], title=key + " filled")
    #else:
    #    myshow(mask2[key] - mask1[key], title=key)
    # Create soft mask
    mask2["soft_" + key] = SoftMask(mask2[key], 4.0) # create soft masks
    #if key == 'cerebellum':
    #    myshow3d(sitk.LabelOverlay(mask2["soft_" + key], mask1[key]+mask2[key]), zslices=[96], title=key)
    #else:
    #    myshow(sitk.LabelOverlay(mask2["soft_" + key], mask1[key]+mask2[key]), title=key)
# Create a third label for watersheds from T1 image
z = 104
t1 = sitk.ReadImage(osjoin(experimentDir, 'newAtlas', 'template_t1.nii.gz'))
t1 = sitk.Resample(t1, interpolator=sitk.sitkLinear,
                   size=t1.GetSize(),
                   outputOrigin=out3['air'].GetOrigin(),
                   outputDirection=out3['air'].GetDirection())
distFile = osjoin(experimentDir, 'Results', 'distanceMap.nii.gz')
skullFile = osjoin(experimentDir, 'Results', 'masks_skull.nii.gz')
if os.path.exists(distFile) and os.path.exists(skullFile):
    dist_img = sitk.ReadImage(distFile)
    mask2['skull'] = sitk.ReadImage(skullFile)
    labels =  mask2['cerebellum'] + (mask2['cerebrum']*2) + mask2['skull']*3
else:
    # myshow(sitk.LabelOverlay(out3['wm'], mask2['cerebrum']+mask2['cerebellum']*2)[:,:,z], "WM masked")
    # myshow(sitk.LabelOverlay(out3['crblwm'], mask2['cerebrum']+mask2['cerebellum']*2)[:,:,z], "CRBLWM masked")
    combined = (mask2['cerebellum'] + mask2['cerebrum']) > 0
    safety_margin = 3
    # myshow(sitk.BinaryDilate(combined, safety_margin) - combined)
    mask2['skull'] = (t1>50) * (1 - sitk.BinaryDilate(combined, safety_margin))
    myshow(mask2['skull'], "Skull mask")
    labels =  mask2['cerebellum'] + (mask2['cerebrum']*2) + mask2['skull']*3
    # Move over the cerebellum and cerebrum values for gm
    dist_img = sitk.SignedMaurerDistanceMap(labels!=0, insideIsPositive=False, squaredDistance=False, useImageSpacing=False)
    myshow(dist_img[:,:,z])
    sitk.WriteImage(dist_img, distFile)
    dumpImageMap(mask2, osjoin(experimentDir, 'Results'), prefix="masks", sep="_")
if debug:
    checkStats(mask2)
mask1 = None

gm_combined = (out3['surfgm'] + out3['crblgm'])
wsd_img = sitk.MorphologicalWatershedFromMarkers(gm_combined * dist_img, labels, markWatershedLine=False)
myshow(wsd_img[:,:,z], "Watershed")

mask3 = {} # Subject-specific masks included
mask3['soft_cerebellum'] = SoftMask(wsd_img==1, 2.0) # smaller smoothing due to label proximity
mask3['soft_cerebrum'] = SoftMask(wsd_img==2, 2.0)
myshow(mask3['soft_cerebrum'], "Soft cerebrum")
myshow(mask3['soft_cerebellum'][:,:,96], "Soft cerebellum")
for key in mask2.keys():
    if key in mask3:
        continue
    else:
        mask3[key] = mask2[key]
if debug:
    checkStats(mask3)
mask2 = None
```

## Gaussian Mixture Model of Cerebellum ##

The cerebellum has very crisp, segregated prior probabilities which is __NOT__ anatomically accurate.  The gray and white matter priors should be more mixed than they are currently.  To address this problem:

1. Create a cerebellum mask
  1. Sum gray and white matter from 'soft' cerebellum results
  2. Create a binary mask at 0.001
  3. Clean up mask (hole filling, closing, etc.)
2. Blur gray and white matter with a Gaussian that has a large size (roughly the size of the folis)
3. Mask the blurred images with the mask
4. Scale the output s.t. the values are > 0.0
5. Linear scale the values s.t. 

$$
GM + WM = \gamma(G(GM) + G(WM))
$$
where G() is the gaussian function

```python
def gammaScale(img1, img2, variance):
    mass = img1 + img2
    assert sitk.Statistics(mass)['Minimum'] == abs(sitk.Statistics(mass)['Minimum'])
    img1_g = sitk.DiscreteGaussian(img1, variance)
    img2_g = sitk.DiscreteGaussian(img2, variance)
    # Bring up values to > 0.0
    min1 = sitk.Statistics(img1_g)['Minimum']
    img1_g = img1_g - min1
    min2 = sitk.Statistics(img2_g)['Minimum']
    img2_g = img2_g - min2
    summed = img1_g + img2_g
    assert sitk.Statistics(summed)['Minimum'] == abs(sitk.Statistics(summed)['Minimum'])
    scale = mass / summed
    return img1_g * scale, img2_g * scale

# z=96
level=10
cerebellum_combined = out4['crblgm'] + out4['crblwm']
myshow(out4['crblgm'][:,:,z])
myshow(out4['crblwm'][:,:,z])
crblgm, crblwm = gammaScale(out4['crblgm'], out4['crblwm'], level)

out5 = {}
out5['crblgm'] = sitk.Mask(crblgm, wsd_img==1)
out5['crblwm'] = sitk.Mask(crblwm, wsd_img==1)
myshow(out5['crblgm'][:,:,z])
myshow(out5['crblwm'][:,:,z])

for key in out4.keys():
    if key in out5.keys():
        continue
    else:
        out5[key] = out4[key]
if debug:
    checkStats(out5, checks=["min"])
out4 = None
```

## Normalize the images ##

```python
def normalizeImages(d, verify=True, tolerance=1e-5):
    """
    Takes in a list of images and normalizes them s.t. the voxels in the sum of the images is rougly equivalent to 1
    """
    mag = sumOverKeys(d, d.keys())  # Magnitude image
    # Normalize by the magnitude image
    n_image_list = []
    count = 0
    dout = {}
    verifyImage(mag, "magnitude", tolerance)
    for key, image in d.items():
        n_img = sitk.Cast(image, sitk.sitkFloat32) / mag  ### <--- How to handle over/underflow?
        if verify: verifyImage(n_img, key, tolerance)
        dout[key] = n_img
    if verify:
        testNormalizeImages(mag, imageList(dout, dout.keys()), tolerance=tolerance)
    return dout

def verifyImage(image, image_key, tolerance):
    c_stats = sitk.Statistics(image)
    for key, value in c_stats.items():
        if isnan(value):
            print "%s is NaN for image %s" % (key, image_key)
            raise ValueError
        if isinf(value):
            print "%s is Inf for image %s" % (key, image_key)
            raise ValueError
        if key == 'Minimum' and value < 0.0:
            print "Minimum is negative for image {0}".format(image_key)
        if key == "Maximum" and value > 1.0 + tolerance:
            print "Maximum is greater than 1.0 for image {0}".format(image_key)
    
def testNormalizeImages(mag, n_image_list, tolerance=1e-5):
    n_mag = sitk.NaryAdd(n_image_list)
    # Verify the normalization
    m_stats = sitk.Statistics(mag)
    n_stats = sitk.Statistics(n_mag)
    assert n_stats["Variance"] <= m_stats["Variance"], "Variance in image sum was not reduced! Original: %f\t Normal: %f" % (n_stats["Variance"], m_stats["Variance"])
    assert n_stats["Maximum"] < 1.0 + tolerance, "Maximum summed value is outside tolerance: %f" % n_stats["Maximum"]
    assert n_stats["Minimum"] > 1.0 - tolerance, "Minimum summed value is outside tolerance: %f" % n_stats["Minimum"]

out7 = normalizeImages(out6, verify=True)
if debug:
    checkStats(out7)
myshow(sumOverKeys(out6, out6.keys()), "Pre-Normalized Sum")
myshow(sumOverKeys(out7, out7.keys()), "Normalized Sum")
out6 = None
```
