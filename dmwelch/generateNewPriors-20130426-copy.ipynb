```python
sys.path.insert(1, '/ipldev/sharedopt/ipython_notebooks/SimpleITK-Notebooks')
import myshow as sitk_show
def show(img, **kwds):
    size = img.GetSize()
    if kwds == {}:
        zslices=range(size[2]/10,size[2]-size[2]/10,size[2]/10)
        sitk_show.myshow3d(img, xslices=[], yslices=[], zslices=zslices, dpi=30)
        return 0
    for key, value in kwds.items():
        if key.lower() == 'x':
            if value == []:
                xslices = range(50,size[0]-50,10)
            else: 
                xslices = value
        elif key.lower() == 'y':
            if value == []:
                yslices=range(50,size[1]-50,20)
            else: 
                yslices = value
        elif key.lower() == 'z':
            if value == []:
                zslices=range(120,size[2]-120,20)
            else:
                zslices = value
        else:
            raise Exception("Unknown dimension")
    sitk_show.myshow3d(img, xslices=xslices, yslices=yslices, zslices=zslices, dpi=30)
    return 0
```

```python
def filesplit(filename):
    """ 
    Split filename by the first '.' in order to append to filename (not extension)
    """
    flist = filename.split('.')
    if len(flist) == 2:
        return flist[0], '.' + flist[1]
    else:
        return flist[0], '.' + '.'.join(flist[1:])
```

```python
def resampleImage(infile, outfile, resolution=(1.0, 1.0, 1.0), interpolator=sitk.sitkLinear):
    image = sitk.ReadImage(infile)
    outImage = image
    old_size = list(image.GetSize())
    old_res = list(image.GetSpacing())
    new_res = list(resolution)
    new_size = [ int((old_size[i] * old_res[i]) / new_res[i]) for i in range(len(new_res))]
    outImage = sitk.Resample(outImage,
                             transform=sitk.Transform(),
                             outputSpacing=resolution,
                             interpolator=interpolator,
                             size=tuple(new_size),
                             outputOrigin=image.GetOrigin(),
                             outputDirection=image.GetDirection())
    felements = outfile.split('.')
    if len(felements) == 1:
        outfile += '.nii.gz'
    outfile = os.path.join(os.getcwd(), outfile)
    sitk.WriteImage(outImage, outfile)
    return outfile

resampleImage(brainMask, outputMask, (0.75,)*3, sitk.sitkNearestNeighbor)
resampleImage(averageT1, resampledT1, (0.75,)*3, sitk.sitkNearestNeighbor)
```

We want to do several steps:

1. Threshold the probability images at 0.1
1. Mask the TISSUE files with a ICV or Brain label 
1. Dilate the mask by a width of 3 mm  
1. Blur the mask by a Gaussian with a kernel twice the width of the dilation.  
1. The TISSUE image will be multiplied by the blurred mask, and 
1. the NOT_TISSUE image will be multiplied by (1 - blurred_mask).  

This should allow us to iterate over all voxels of overlap to verify that P(tissue) + P(not) = 1.0

```python
def createNormalizedImage(directory):
    """ Normalize so all voxels are between 0.0 and 1.0 """ 
    listdir = os.listdir(directory)
    filelist = []
    for item in listdir:
        if item[-5:] == '.nrrd':
            filelist.append(item)
    for index in range(len(filelist)):
        print filelist[index]
        try:
            filename = os.path.join(directory, filelist[index])
            print filename
            image = sitk.ReadImage(filename)
        except:
            print os.path.join(directory, filelist[index])
            raise
        #if index == 0:
        #    summed = sitk.Cast(image, sitk.sitkInt32)
        #    summed *= 0
        maximum = sitk.Statistics(image).GetBasicMeasurementMap()['Maximum']
        minimum = sitk.Statistics(image).GetBasicMeasurementMap()['Minimum']
        # print maximum
        image = (image - minimum) / (maximum - minimum)
        castImage = sitk.Cast(image * 100.0, sitk.sitkInt32)
        sitk.WriteImage(castImage, os.path.join(directory, 'normed/normed_' + filelist[index])) 
    # sitk.WriteImage(summed, os.path.join(directory, 'total.nii.gz'))
```
