```python
import copy, csv, os
import subprocess as sp

class Concater(object):
    def __init__(self, *args, **kwds):
        keys = ['grant', 'date', 'outDir']
        keys.reverse()
        args = list(args)
        for key in keys:
            value = kwds.pop(key, None)
            if value is None:
                value = args.pop()
            self.__setattr__(key, value)
        super(Concater, self).__init__()
        if self.grant.lower() == 'predict':
            self.__predict__()
        elif self.grant.lower() == 'trackon':
            self.__trackon__()
        else:
            raise ValueError('self.grant value is: "%s".\nAllowed values are ["predict", "trackon"] (case-insensitive)' % self.grant) 
        self._formatDate()
        self._constructOutputDirectory()
        self.qaFile = None
        self._constructQAFilePath()
        
            
    def __trackon__(self):
        self._baseDir = '/hjohnson/TrackOn'
        self._qaFileName = 'track_dwi_raw_qc.csv'
        if self.outDir is None:
            self.outDir = '20130211'
        
        
    def __predict__(self):
        self.baseDir = '/paulsen/MRx'
        self._qaFileName = 'predict_dwi_raw_qc.csv'
        if self.outDir is None:
            self.outDir = '20130211'
    
    def _formatDate(self):
        date = self.date.split('-')
        if len(date) == 1:
            self.date = '%s-%s-%s' % (self.date[:4], self.date[4:6], self.date[6:8])
    
    def _constructOutputDirectory(self):
        if not os.path.exists(self.outDir):
            experimentDir = os.path.join(self._baseDir, 'Experiments')
            self.outDirName = os.path.join(experimentDir, self.outDir)
            assert os.path.exists(self.outDirName), \
                   'Cannot find output directory in "%s": %s' % (experimentDir, self.outDir)
        else:
            self.outDirName = copy.deepcopy(self.outDir)
        
    
    def _constructQAFilePath(self):
        if self.qaFile is None:
            self.qaFile = os.path.join(self._baseDir, 'reports', self.date, self._qaFileName)
            assert os.path.exists(self.qaFile), "QA .csv file cannot be found!"
        
    
    def readQAFile(self):
        with open(self.qaFile, 'rb') as csvfile:
            self._fileList = csvfile.readlines()
        self.readCSV()
            
            
    def readCSV(self):
        reader = csv.reader(self._fileList)
        previousDir = ""
        mergeList = []
        for currentPath in reader: # only one: filename
            if previousDir == "":
                previousDir, currentFile = os.path.split(currentPath[0])
                mergeList.append(currentPath[0])
                #print "Filling previousDir: ", previousDir
                #print "File: ", currentFile
            else:
                currentDir, currentFile = os.path.split(currentPath[0])
                if currentDir == previousDir:
                    mergeList.append(currentPath[0])
                    #print "Same session: ", currentDir
                    #print "File: ", currentFile
                else:
                    self._findConcatScans(mergeList)
                    mergeList = [currentPath[0]]
                    previousDir = currentDir
                    #print "Different session: ", currentDir
                    #print "File: ", currentFile
                    
                    
    def _findConcatScans(self, mergeList):
        pathElements = list(mergeList[0].split(os.path.sep))
        fileName = pathElements.pop() 
        container = pathElements.pop() # ANONRAW
        session = pathElements.pop()
        subject = pathElements.pop()
        site = pathElements.pop()
        # base = os.path.sep + os.path.join(*pathElements)
        destinationDir = os.path.join(self.outDirName, site, subject, session)
        if not os.path.isdir(destinationDir):
            os.makedirs(destinationDir, mode=0755)
        gradientCheck = ""
        inputs = []
        for fullPath in mergeList:
            _, fileName = os.path.split(fullPath)
            _, gradInfo = fileName.split('-')
            gradient, _ = gradInfo.split('_')
            if gradientCheck == "":
                gradientCheck = gradient
                inputs.append(fullPath)
            elif gradientCheck == gradient:
                inputs.append(fullPath)
            else:
                #self.runGTractConcatDWI(inputs, destinationDir)
                print "Here"
                print inputs, destinationDir
                inputs = [fullPath]
                
        self._runGTractConcatDWI(inputs, destinationDir)
            
                    
    def _runGTractConcatDWI(self, inputs, destinationDir):
        executable = '/IPLlinux/ipldev/scratch/welchdm/bld/BSA-gtrackConcatDWI/bin/gtractConcatDwi'
        _, fileName = os.path.split(inputs[0])
        prefix, suffix = fileName.split('-')
        gradient, _ = suffix.split('_')
        outFileName = prefix + "-" + gradient + "_" + "concat.nrrd"
        inargs = []
        for index in range(len(inputs)):
            inargs.append('--inputVolume')
            inargs.append(inputs[index])
        argList = [executable] + inargs + ['--outputVolume', os.path.join(destinationDir, outFileName)]
        
        gTract = sp.call(argList)
        if gTract != 0:
            print "Concatenation failed for:\n'%s'" % " ".join(argList)
        
    def __str__(self):
        string = ""
        for key, value in self.__dict__.items():
            if key == '_fileList':
                string += "_fileList: list({0}, ...)\n".format(self._fileList[0])
            else:
                string += "{0}: {1}\n".format(key, value)
        return string
        
```
