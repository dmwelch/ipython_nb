```python
def binaryBoundingBoxMask(image, returnPoints=False):
    """ Given a binary image, return a binary bounding box mask image in the same space """
    import SimpleITK as sitk
    import numpy as np
    import pinc
    array = sitk.GetArrayFromImage(image)
    non_zero = np.nonzero(array)
    extremes = []
    for dd in range(array.ndim):
        extremes.append((np.min(non_zero[dd]), np.max(non_zero[dd])))
    # Create the binary mask
    evalString = ""
    for ii in range(array.ndim):
        evalString += "np.arange(*extremes[{0}]), ".format(ii)
    evalString = evalString[:-2]
    indicies = eval("np.ix_(" + evalString + ")")
    array[indicies] = array.max()
    output = sitk.GetImageFromArray(array)
    output.CopyInformation(image)
    if True:
        pinc.vDisplayCenterSlice([output-image], ['Double-check'], 0.5)
    # Create 
    if returnPoints:
        pt1 = []
        pt2 = []
        for jj in range(len(extremes)):
            # DON'T FORGET TO CAST NUMPY TYPES TO PYTHON INT
            pt1.append(int(extremes[jj][0]))
            pt2.append(int(extremes[jj][1]) + 1)
        return output, (tuple(pt1), tuple(pt2))
    return output
```

```python
def ComputeBoundingBox(img):
    """ Bradley Lowenkamp version of binaryBoundingBoxMask, only returns extrema points """
    npImg  = sitk.GetArrayFromImage( img )
    idxs = np.argwhere( npImg )
    return idxs.min(0), idxs.max(0) + 1
```
