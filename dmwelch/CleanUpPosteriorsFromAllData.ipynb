```python
import SimpleITK as sitk
import os
import sys
def DisplayCenterSlice(image, title="",percentage=0.50):
    """ Display the slice of an sitkImage that is % throught the volume"""
    size=list(image.GetSize())
    dims=image.GetDimension()
    if dims == 2:
        index=[0,0]
    else:
       index=[0,0,int(size[2]*percentage)]
       size[2]=0
    twoDImage=sitk.Extract(image,size,index)
    np_image=sitk.GetArrayFromImage(twoDImage)
    myplot=plt.imshow(np_image, cmap=cm.gray, interpolation=None)
    myplot.set_label("T2_image")
    plt.title(title)
    plt.show()
    return myplot

def add_all_images(image_list):
    accum=sitk.Image(image_list[0])
    for ii in image_list[1:]:
        accum += ii
    return accum
```

```python
brain=sitk.ReadImage('/Users/johnsonhj/Dropbox/DATA/BLOOD_ADD/template_brain.nii.gz')
new_brain=sitk.ReadImage('/Users/johnsonhj/Dropbox/DATA/BLOOD_ADD/template_brain_wblood.nii.gz')
old_ABC_labels_image=sitk.ReadImage('/Users/johnsonhj/src/BSA-clang31/ReferenceAtlas-build/Atlas/Atlas_20121118/template_ABC_lables.nii.gz')
```

```python
old_ABC_blood=sitk.BinaryThreshold(old_ABC_labels_image,5,5)
xorblood=sitk.Or(old_ABC_blood, sitk.Xor(brain,new_brain))
fuzzyxorblood=sitk.DiscreteGaussian(sitk.Cast(xorblood,sitk.sitkFloat32),5.0)
stats_value=sitk.Statistics(fuzzyxorblood)
image_max=stats_value.GetVectorOfMeasurementValues()[0]
notfuzzyxorblood=1.0-(fuzzyxorblood/image_max)
```

```python
DisplayCenterSlice(xorblood,'xorblood',.36)
DisplayCenterSlice(fuzzyxorblood,'fuzzyxorblood',.36)
DisplayCenterSlice(notfuzzyxorblood,'notfuzzyxorblood',.36)

#sitk.Show(fuzzyxorblood*255.0,'fuzzyxorblood.nii')
```

```python

#t1_image=sitk.ReadImage(t1_images_fn[0])
#DisplayCenterSlice(old_ABC_labels_image,'old_labels')
#DisplayCenterSlice(old_ABC_labels_image,'old_labels')

post_images=dict()
for pp in post_images_fn:
    bnname=os.path.basename(pp).replace('.nii.gz','')
    post_images[bnname]=sitk.ReadImage(pp)
    #DisplayCenterSlice(post_images[bnname],bnname)
    
struct_images=dict()
for pp in struct_images_fn:
    bnname=os.path.basename(pp).replace('_trans.nii.gz','').replace('subjectANNLabel_','')
    struct_images[bnname]=sitk.ReadImage(pp)
    #DisplayCenterSlice(struct_images[bnname],bnname)
```

```python
vb=post_images['POSTERIOR_VB_trans']
notvb=post_images['POSTERIOR_NOTVB_trans']
high_prob_vb=sitk.BinaryThreshold(vb,0.33)
high_prob_notvb=sitk.BinaryThreshold(notvb,0.75)

#DisplayCenterSlice(high_prob_notvb+high_prob_vb,'vb',.45)
Preservere_VB_Region=sitk.Cast(high_prob_notvb+high_prob_vb,sitk.sitkFloat32)
NOT_Preservere_VB_Region=sitk.Cast(1.0-Preservere_VB_Region,sitk.sitkFloat32)
new_not_vb=notvb*0.5*NOT_Preservere_VB_Region+notvb*Preservere_VB_Region
DisplayCenterSlice(new_not_vb,'new_not_vb',.45)

post_images['POSTERIOR_NOTVB_trans']=new_not_vb

finalnotvb=post_images['POSTERIOR_NOTVB_trans']
finalvb=post_images['POSTERIOR_VB_trans']

move_to_vb=sitk.Cast(new_brain,sitk.sitkFloat32)*finalnotvb
finalvb=sitk.DiscreteGaussian(finalvb+move_to_vb,1.5)
finalnotvb=sitk.DiscreteGaussian(finalnotvb*(1.0 - sitk.Cast(new_brain,sitk.sitkFloat32)),1.5)*0.85

post_images['POSTERIOR_NOTVB_trans']=finalnotvb
post_images['POSTERIOR_VB_trans']=finalvb

accum_non_air=add_all_images(post_images.values())
new_air=1-accum_non_air
DisplayCenterSlice(new_air,'NonAirAccumulator')
post_images['POSTERIOR_AIR_trans']=new_air

```

```python
post_images['POSTERIOR_ACCUMBEN_trans']=post_images['POSTERIOR_ACCUMBEN_trans']+struct_images['l_accumben']+struct_images['r_accumben']
post_images['POSTERIOR_HIPPOCAMPUS_trans']=post_images['POSTERIOR_HIPPOCAMPUS_trans']+struct_images['l_hippocampus']+struct_images['r_hippocampus']
post_images['POSTERIOR_THALAMUS_trans']=post_images['POSTERIOR_THALAMUS_trans']+struct_images['l_thalamus']+struct_images['r_thalamus']
post_images['POSTERIOR_GLOBUS_trans']=post_images['POSTERIOR_GLOBUS_trans']+struct_images['l_globus']+struct_images['r_globus']
post_images['POSTERIOR_PUTAMEN_trans']=post_images['POSTERIOR_PUTAMEN_trans']+struct_images['l_putamen']+struct_images['r_putamen']
post_images['POSTERIOR_CAUDATE_trans']=post_images['POSTERIOR_CAUDATE_trans']+struct_images['l_caudate']+struct_images['r_caudate']
```
