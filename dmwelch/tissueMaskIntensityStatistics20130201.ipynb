```python
CSVFileName = '/ipldev/scratch/eunyokim/tissueMaskStats.csv'
basedir = '/paulsen/Experiments/20130202_PREDICTHD_Results'

import csv
import os
import re
import sys
sys.path.insert(1, '/ipldev/scratch/welchdm/bld/rs-fMRI/BSA/SimpleITK-build/lib/')
sys.path.insert(2, '/ipldev/scratch/welchdm/bld/rs-fMRI/BSA/SimpleITK-build/bin/')
sys.path.insert(3, '/ipldev/scratch/welchdm/bld/rs-fMRI/BSA/SimpleITK-build/Wrapping/')
sys.path.insert(4, '/ipldev/sharedopt/ipython_notebooks/SimpleITK-Notebooks/')
sys.path.insert(5, '/ipldev/sharedopt/ipython_notebooks/common/')
import SimpleITK as sitk
from myshow import *
```

```python
csvFile = open(CSVFileName, 'w')
headings = ['experiment', 'site', 'subject', 'session', 
            'WM_mean(T1)(1)', 'WM_sigma(T1)(1)', 'WM_mean(T2)(1)', 'WM_sigma(T2)(1)', 
            'WM_mean(T1)(2)', 'WM_sigma(T1)(2)', 'WM_mean(T2)(2)', 'WM_sigma(T2)(2)', 
            'GM_mean(T1)(1)', 'GM_sigma(T1)(1)', 'GM_mean(T2)(1)', 'GM_sigma(T2)(1)', 
            'GM_mean(T1)(2)', 'GM_sigma(T1)(2)', 'GM_mean(T2)(2)', 'GM_sigma(T2)(2)', 
            'CSF_mean(T1)(1)', 'CSF_sigma(T1)(1)', 'CSF_mean(T2)(1)', 'CSF_sigma(T2)(1)', 
            'CSF_mean(T1)(2)', 'CSF_sigma(T1)(2)', 'CSF_mean(T2)(2)', 'CSF_sigma(T2)(2)']

dWriter = csv.DictWriter(csvFile, headings,
                         restval='', extrasaction='raise', dialect='excel')
dWriter.writeheader()
csvFile.close()

for path, _, files in os.walk(basedir):
    if "POSTERIOR_WM.nii.gz" in files:
        csvFile = open(CSVFileName, 'a')
        dWriter = csv.DictWriter(csvFile, headings, restval='', extrasaction='raise', dialect='excel')
        wmImage1 = sitk.ReadImage(os.path.join(path, "POSTERIOR_WM.nii.gz"))
        wmMask = sitk.BinaryThreshold(wmImage1, 0.75, 1.01, 1, 0)
        
        wmImage2 = sitk.ReadImage(os.path.join(path, "POSTERIOR_CRBLWM.nii.gz"))
        wmMask += sitk.BinaryThreshold(wmImage2, 0.75, 1.01, 1, 0)
        
        wmClose1 = sitk.ClosingByReconstruction(wmMask, 1)
        wmClose2 = sitk.ClosingByReconstruction(wmMask, 2)
        
        gmImage1 = sitk.ReadImage(os.path.join(path, "POSTERIOR_SURFGM.nii.gz"))
        gmMask = sitk.BinaryThreshold(gmImage1, 0.75, 1.01, 1, 0)
        
        gmImage2 = sitk.ReadImage(os.path.join(path, "POSTERIOR_CRBLGM.nii.gz"))
        gmMask += sitk.BinaryThreshold(gmImage2, 0.75, 1.01, 1, 0)
        
        gmClose1 = sitk.ClosingByReconstruction(gmMask, 1)
        gmClose2 = sitk.ClosingByReconstruction(gmMask, 2)
        
        csfImage = sitk.ReadImage(os.path.join(path, "POSTERIOR_CSF.nii.gz"))
        csfMask = sitk.BinaryThreshold(csfImage, 0.75, 1.01, 1, 0)
        
        csfClose1 = sitk.ClosingByReconstruction(csfMask, 1)
        csfClose2 = sitk.ClosingByReconstruction(csfMask, 2)
        
        t1File = os.path.join(path, 't1_average_BRAINSABC.nii.gz')
        t2File = os.path.join(path, 't2_average_BRAINSABC.nii.gz')
        out = dict()
        parts = path.split(os.path.sep)
        _ = parts.pop()
        session = parts.pop()
        subject = parts.pop()
        site = parts.pop()
        experiment = parts.pop()
        measurements = [experiment, site, subject, session]
        for mask in [wmClose1, wmClose2, gmClose1, gmClose2, csfClose1, csfClose2]:
            stats = sitk.LabelStatisticsImageFilter()
            stats.Execute(sitk.ReadImage(t1File), mask)
            # maskedImage = sitk.Mask(sitk.ReadImage(t1File), mask)
            # stats = sitk.Statistics(maskedImage)
            # measures = stats.GetBasicMeasurementMap()
            t1Mean = stats.GetMean(1) # measures['Mean']
            t1Sigma = stats.GetSigma(1) # measures['Sigma']
            measurements.append(t1Mean)
            measurements.append(t1Sigma)
            try:
                stats = sitk.LabelStatisticsImageFilter()
                stats.Execute(sitk.ReadImage(t2File), mask)
                # maskedImage = sitk.Mask(sitk.ReadImage(t2File), mask)
                # stats = sitk.Statistics(maskedImage)
                # measures = stats.GetBasicMeasurementMap()
                t2Mean = stats.GetMean(1) # measures['Mean']
                t2Sigma = stats.GetSigma(1) # measures['Sigma']
            except RuntimeError, err:
                t2Mean = None
                t2Sigma = None
            measurements.append(t2Mean)
            measurements.append(t2Sigma)
        for key, value in zip(headings, measurements):
            out[key] = value
        print "Writing: ", out
        dWriter.writerow(out)
        csvFile.close()

```
