```python
import os
import sys

import SimpleITK as sitk
import pinc

notebookDir = os.path.dirname(os.getcwd())
experimentDir = os.path.join('/paulsen', 'Experiments', '20130426-generateNewPriors')
```

# Create masks #

1. Whole brain mask
2. Brainstem
3. Cerebellum
4. Cerebrum

1 mm atlas is on the 'critical path', so we will only work on those. (0.75mm is marginally more work)

## Whole brain ##

```python
atlasFile = os.path.join(experimentDir, 'newAtlas', 'atlas.nii.gz')
atlas = sitk.ReadImage(atlasFile)
```

```python
array = sitk.GetArrayFromImage(atlas)
labels = numpy.unique(array)
print "Number of labels:", labels.size
```

```python
wholeBrain = (atlas < 4000) + (atlas >= 1) + (atlas != 77) + (atlas != 215) + (atlas != 85) == 5
wholeBrain = sitk.Cast(wholeBrain, sitk.sitkUInt8)
sitk.WriteImage(wholeBrain, os.path.join(experimentDir, 'newAtlas', 'brainmask.nii.gz'))
pinc.show(wholeBrain)
```

```python
wb_closed = sitk.BinaryMorphologicalClosing(wholeBrain, 9)
pinc.show(wholeBrain - wb_closed)
```

```python
wb_dilated = sitk.BinaryDilate(wb_closed, int(3.0 / wb_closed.GetSpacing()[0])) # 0.75 mm * 4 = 3 mm
pinc.show(wb_dilated - wholeBrain)
```

## Brainstem ##

```python
labels = [15, 16, 19, 35, 40, 61, 66, 71, 79, 84]
brainstem = atlas * 0
brainstem = sitk.Cast(brainstem, sitk.sitkUInt8)
for label in labels:
    temp = (atlas == label)
    brainstem = sitk.Add(brainstem, sitk.Cast(temp, brainstem.GetPixelIDValue()))
sitk.WriteImage(brainstem, os.path.join(experimentDir, 'newAtlas', 'brainstem.nii.gz'))
pinc.show(brainstem)
```

##### For brainstem at 0.75mm #####

```python
# There is a small hole in the atlas label @ 0.75 mm, so we need to close it...
#
# DO NOT DO THIS ----> bs_closed = sitk.BinaryMorphologicalClosing(brainstem, 2)
# Instead, find the indicies and set equal to 1, thereby keeping edges correct!
# 
for x in range(169, 174):
    for y in range(227, 230):
        for z in range(103, 107):
            brainstem[x, y, z] = 1
pinc.show(brainstem)
sitk.WriteImage(brainstem, os.path.join(experimentDir, 'newAtlas', 'brainstem_075mm.nii.gz'))
```

## Cerebellum ##

```python
cerebellum = (atlas > 299) + (atlas < 380) + (atlas == 7) + (atlas == 46) + (atlas == 80) != 1
cerebellum = sitk.Cast(cerebellum, sitk.sitkUInt8)
sitk.WriteImage(cerebellum, os.path.join(experimentDir, 'newAtlas', 'cerebellum.nii.gz'))
pinc.show(cerebellum)
```

```python
cerebellum = sitk.ReadImage(os.path.join(experimentDir, 'newAtlas', 'cerebellum.nii.gz'))
```

```python
cbl_closed = sitk.BinaryMorphologicalClosing(cerebellum, 3)
pinc.show(cerebellum - cbl_closed)
```

```python
cbl_dilated = sitk.BinaryDilate(cbl_closed, int(3.0 / cbl_closed.GetSpacing()[0]))
pinc.show(cbl_dilated - cerebellum)
sitk.WriteImage(cbl_dilated, os.path.join(experimentDir, 'cerebellum_dilated.nii.gz'))
```

## Cerebrum ##

```python
cerebrum = wholeBrain - cerebellum - brainstem
sitk.WriteImage(cerebrum, os.path.join(experimentDir, 'newAtlas', 'cerebrum.nii.gz'))
pinc.show(cerebrum)
```

```python
cerebrum = sitk.ReadImage(os.path.join(experimentDir, 'newAtlas', 'cerebrum.nii.gz'))
```

```python
cbr_closed = sitk.BinaryMorphologicalClosing(cerebrum, 3)
pinc.show(cerebrum - cbr_closed)
```

```python
cbr_dilated = sitk.BinaryDilate(cbr_closed, int(3.0 / cbr_closed.GetSpacing()[0]))
pinc.show(cbr_dilated - cerebrum)
sitk.WriteImage(cbr_dilated, os.path.join(experimentDir, 'cerebrum_dilated.nii.gz'))
```
