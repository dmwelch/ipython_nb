```python
import os.path
import SimpleITK as sitk
import numpy as np
labelDirectory = 'SingleRFSegmentations'
```

```python
def formatLabel(label):
    """
    Assumes that the label can be split by the '_' character.
    """
    side, anatomy = label.split('_')
    if side.lower() in ['l', 'left']:
        side = 'left'
    elif side.lower() in ['r', 'right']:
        side = 'right'
    else:
        raise ValueError('Label %s is not recognized: cannot determine side %s' % (label, side)) 
    label = '_'.join([side,anatomy])
    return label
```

```python
def calculateLabelVolume(dirname, label):
    labelFile = os.path.join(dirname, labelDirectory, label + '_seg.nii.gz')
    # print labelFile
    assert os.path.exists(labelFile)
    image = sitk.ReadImage(labelFile)
    nda = sitk.GetArrayFromImage(image)
    maskSum = nda.sum()
    print binarySum
    size = image.GetSpacing()
    print size
    return maskSum * size[0] * size[1] * size[2]
```

```python
def getVolume(args=[], kwds={}):
    dirname = labels = project = subject = session = experimentDir = None
    experimentDir = config['experimentDir'] ### HACK
    for key, value in kwds:
        if key == 'dirname':
            dirname = check_file(value)
        elif key == 'labels':
            labels = value
        elif key == 'project':
            project = value
        elif key == 'subject':
            subject = value
        elif key == 'session':
            session = value
    if session is None:
        session = args.pop()
    if subject is None:
        subject = args.pop()
    if project is None:
        project = args.pop()
    if labels is None:
        labels = args
    if dirname is None:
        try:
        # config needs to be accessible
            dirname = os.path.join(experimentDir, project, subject, session)
        except Exception, err:
            raise err
    ### DEBUGGING ###
    #print labels
    #print dirname
    #print project
    #print subject
    #print session
    ### END DEBUG ###
    # labels = map(formatLabel, labels) # convert shorthand to human-readable
    volume = 0.0
    for label in labels:
        volume += calculateLabelVolume(dirname, label)
    return volume
    
```

```python
import os.path
import SimpleITK as sitk
import numpy as np
from warnings import warn
labelDirectory = 'TissueClassify'
tolerance = [0.51, 1.01]
white_matter_tol = [0.99, 1.01]
```

```python
def calculateBinaryVolume(dirname, label):
    labelFile = os.path.join(dirname, labelDirectory, 'POSTERIOR_'+ label.capitalize() + '.nii.gz')
    # print labelFile
    assert os.path.exists(labelFile)
    image = sitk.ReadImage(labelFile)
    if label == WM:
        binary = sitk.BinaryThreshold(image, 0.99, 1.01, 1, 0)
        warn("Using higher lower threshold 0.99 for white matter binary calculations.")
    else:
        binary = sitk.BinaryThreshold(image, 0.51, 1.01, 1, 0)
    nda = sitk.GetArrayFromImage(binary)
    maskSum = nda.sum()
    print binarySum
    size = image.GetSpacing()
    print size
    return maskSum * size[0] * size[1] * size[2]
```

```python
import os
import sys
sys.path.insert(1, '/ipldev/scratch/welchdm/src/BRAINSStandAlone/AutoWorkup/')

import brains
#parentDir = '/hjohnson/TrackOn/Experiments/20130109_TrackOn_Results'

def csfPerBrainMatter(directory):
    csf = brains.metrics.calculateBinaryVolume(directory, 'csf_total')
    gm = brains.metrics.calculateBinaryVolume(directory, 'gm_total')
    # crblwm = brains.metrics.calculateBinaryVolume(directory, 'wm')
    wm = brains.metrics.calculateBinaryVolume(directory, 'wm_total')
    # surfgm = brains.metrics.calculateBinaryVolume(directory, 'surfgm')
    # print "CSF: %f\tGM: %f\tWM: %f"%(csf, gm, wm)
    return csf / (gm + wm)
    
brains.config.loadConfiguration('/ipldev/scratch/welchdm/brains_20130130.config')
count = 1
root = brains.config._config.get('Results', 'directory')
with open('/IPLlinux/ipldev/sharedopt/ipython_notebooks/dmwelch/csfPerSumOfGreyAndWhiteMatter.csv', 'w') as fid:
    fid.write("Experiment, project, subject, session, ratio (CSF/(GM+WM))\n")
    posteriorDir = brains.config._config.get('Results', 'accumulated')
    for experiment, projects, _ in os.walk(root):
        #if count > 10:
        #    break
        if posteriorDir in projects:
            temp = os.path.abspath(experiment)
            ratio = csfPerBrainMatter(temp)
            temp = temp.replace('/hjohnson/TrackOn/Experiments/', '')
            outs = tuple(temp.split(os.path.sep)) + (ratio,)
            fid.write('{0}, {1}, {2}, {3}, {4}\n'.format(*outs))
            # count +=1
```

# Usage: Case 2
* * *
Hans wants the intercranial volume (ICV) for all the files

$$ CSF + WM + GM + VB $$

```python
import os
if os.path.exists('/ipldev/scratch/welchdm/src/BRAINSStandAlone/AutoWorkup/'):
    import sys
    sys.path.insert(1, '/ipldev/scratch/welchdm/src/BRAINSStandAlone/AutoWorkup/')
else:
    raise RuntimeError('Autoworkup directory is not found!')

try:
    import brains
except Exception, err:
    print sys.path
    raise err


def icvVolume(directory):
    csf = brains.metrics.calculateBinaryVolume(directory, 'csf_total')
    gm = brains.metrics.calculateBinaryVolume(directory, 'gm_total')
    wm = brains.metrics.calculateBinaryVolume(directory, 'wm_total')
    vb = brains.metrics.calculateBinaryVolume(directory, 'vb_total')
    return csf + gm + wm + vb

def ventricleVolume(directory):
    return brains.metrics.calculateBinaryVolume(directory, 'csf_total')
    
brains.config.loadConfiguration('/ipldev/scratch/welchdm/brains_20130130.config')
root = brains.config._config.get('Results', 'directory')
with open('/IPLlinux/ipldev/sharedopt/ipython_notebooks/dmwelch/icv.csv', 'w') as fid:
    fid.write("Experiment, project, subject, session, ICV (CSF + GM + WM + VB), Ventricles (CSF)\n")
    posteriorDir = brains.config._config.get('Results', 'accumulated')
    for experiment, projects, _ in os.walk(root):
        if posteriorDir in projects:
            temp = os.path.abspath(experiment)
            total = icvVolume(temp)
            ventricle = ventricleVolume(temp)
            temp = temp.replace('/hjohnson/TrackOn/Experiments/', '')
            outs = tuple(temp.split(os.path.sep)) + (total, ventricle)
            fid.write('{0}, {1}, {2}, {3}, {4}, {5}\n'.format(*outs))
```

```python
import os.path
import sys
sys.path.insert(1, '/ipldev/scratch/welchdm/src/BRAINSStandAlone/AutoWorkup/')

import brains

brains.metrics.loadConfiguration('/ipldev/scratch/welchdm/cfg/BRAINSStandAlone/brains_20130109.config')
```

```python
root = brains.config._config.get('Results', 'directory')
accDir = brains.config._config.get('Results', 'accumulated')
partDir = brains.config._config.get('Results', 'partials')

# project = 'HDNI_004'
# subject = '156454974'
# session = '156454974_20100716_30'
# brains.metrics.getPosteriorVolume('icv', project, subject, session)

allLabels = brains.metrics.partials + brains.metrics.accumulated + ['icv']
title = "Experiment, project, subject, session"

for label in allLabels:
    title += ", %s" % label.upper()
title += "\n"
with open('/ipldev/scratch/welchdm/allOut.csv', 'w') as fid:
    fid.write(title)

for experiment, projects, _ in os.walk(root):
    if accDir in projects or partDir in projects:
        temp = os.path.abspath(experiment)
        volumes = []
        for label in allLabels:
            volumes.append(brains.metrics.getPosteriorVolume(labels=label, dirname=temp, binary=True))
            
        with open('/ipldev/scratch/welchdm/allOut.csv', 'a') as fid:
            outs = temp.replace(root, os.path.basename(root))
            outs = tuple(outs.split(os.path.sep))
            print "outs:", outs
            fid.write('{0}, {1}, {2}, {3}'.format(*outs))
            
            for index in range(len(volumes)):
                fid.write(", {0}".format(volumes[index]))
            fid.write("\n")
```
